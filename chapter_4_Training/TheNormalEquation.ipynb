# The Normal Equation
To find the value of θ that minimizes the cost function, there is a closed-form solution
—in other words, a mathematical equation that gives the result directly. This is called
the Normal Equation 

* My point of views: The Normal Equation is a function to find $\theta$ that minizes the cost function


$$Equation\ 4.4\ Normal\ Equation\$$

```python
import numpy as np
from data_set import get_x_y
```

```python
X, y = get_x_y()
X.shape
```

```python
import matplotlib.pyplot as plt
# s = np.random.rand(*X.shape) * 800 + 500
plt.scatter(X, y, s, c="g", alpha=0.5, marker=r'$chê$')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X]    # add x0 = 1 to each instance
X_b.shape
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.scatter(X, y, c="g", alpha=0.5, marker=r'o')
plt.xlabel('X')
plt.ylabel('y')
plt.plot(X_new, y_predict, "r-")
plt.plot(X_new, y_predict, "r-")
plt.axis([0, 2, 0, 15])
plt.show()
```
