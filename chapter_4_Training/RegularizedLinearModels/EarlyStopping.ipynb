# Early Stopping
Early stopping: stop training as soon as the validation error reaches a minimum

It is such a simple and efficient regularization technique that Geoffrey Hinton called it a “beautiful free lunch.”

<img src="es.png" width=1920 height=1080 />

With Stochastic and Mini-batch Gradient Descent, the curves are
not so smooth, and it may be hard to know whether you have
reached the minimum or not. One solution is to stop only after the
validation error has been above the minimum for some time (when
you are confident that the model will not do any better), then roll
back the model parameters to the point where the validation error
was at a minimum.

```python
import numpy as np
from sklearn import clone
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
```

```python
X = np.random.randn(100,1)
y = 0.5*X**2 +0.5 * X + 10
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size = 0.7)
```

```python
poly_scaler = Pipeline([
    ("polynomial_features", PolynomialFeatures(degree=90, include_bias=False)),
    ("std_scaler", StandardScaler())
])
```

```python
X_train_poly_scale = poly_scaler.fit_transform(X_train)
X_val_poly_scale = poly_scaler.fit_transform(X_val)
```

```python
sgd_reg = SGDRegressor(max_iter=1, warm_start=True, penalty=None,
 learning_rate="constant", eta0=0.0005)
```

```python
minimum_val_error = float("inf")
best_epoch = None
best_model = None
```

```python
for epoch in range(1000):
    sgd_reg.fit(X_train_poly_scale, y_train)
    y_pred = sgd_reg.predict(X_val_poly_scale)
    val_err = mean_squared_error(y_pred, y_val)
    if val_err < minimum_val_error:
        minimum_val_error = val_err
        best_epoch = epoch
        best_model = clone(sgd_reg)
```

Note that with ` warm_start=True `, when the fit() method is called, it just continues
training where it left off instead of restarting from scratch
