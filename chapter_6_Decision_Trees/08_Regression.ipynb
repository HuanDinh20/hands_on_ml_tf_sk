# Regression
Decision Trees are also capable of performing regression tasks.


```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
data = load_iris()
X = data.data[:, 2:]
y = data.target
```

```python
tree_reg = DecisionTreeRegressor(max_depth=2)
tree_reg.fit(X, y)
```

```python
import matplotlib.pyplot as plt
fig  = fig = plt.figure(figsize=(12,8))
plot_tree(tree_reg, filled=True, rounded=True, 
          class_names=data.target_names, 
          feature_names=data.feature_names[2:],)
```

This tree looks very similar to the classification tree you built earlier. The main differ‐
ence is that instead of predicting a class in each node, it predicts a value

Just like for classification tasks, Decision Trees are prone to overfitting when dealing
